```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
fuel_econ = pd.read_csv('/content/drive/MyDrive/Machine Learning BSE/fuel_econ.csv')
fuel_econ.head()
```

```python
binwidth = 60
plt.hist(fuel_econ.co2, edgecolor="black", bins=range(min(fuel_econ.co2), max(fuel_econ.co2) + binwidth, binwidth))
plt.xticks(range(30, max(fuel_econ.co2) + binwidth, binwidth))
plt.xlabel("CO2")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(fuel_econ.drive)
plt.xticks(rotation=60)
plt.ylabel("Frequency")
plt.xlabel("Drive")
plt.show()
```

```python
plt.hist(fuel_econ.cylinders)
plt.ylabel("Frequency")
plt.xlabel("Cylinder")
plt.show()
```

```python
sns.heatmap(fuel_econ.corr(), cmap="RdBu")
plt.xticks(rotation=45, ha='right')
plt.show()
```
