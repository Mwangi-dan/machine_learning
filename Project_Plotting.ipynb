```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
fuel_econ = pd.read_csv('/content/drive/MyDrive/Machine Learning BSE/fuel_econ.csv')
fuel_econ.head(7)
```

```python
plt.hist(fuel_econ['comb'], bins=20, edgecolor='black')
plt.xlabel('Combined Fuel Efficiency')
plt.ylabel('Frequency')
plt.title('Distribution of Combined Fuel Efficiency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

The histogram above displays the combined fuel efficiency across the dataset.

Most of the vehicles in the dataset have a combined fuel efficiency of close to 25.

```python
plt.hist(fuel_econ['co2'], bins=20, edgecolor='black')
plt.xlabel('CO2 Emissions')
plt.ylabel('Frequency')
plt.title('Distribution of CO2 Emissions')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

The most common emmissions of CO2 from vehicles in the dataset is at 320. Fewer vehicles have emissions below 300

```python
numerical_columns = ['displ', 'cylinders', 'city', 'highway', 'comb', 'co2']
corr_matrix = fuel_econ[numerical_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

The heatmap displays the correlation coefficients between city fuel efficiency, cylinders, C02 emmissions, highway fuel efficiency.

Positive correlations indicate a string relationship while negative correleations indicate a negative relatioships.
